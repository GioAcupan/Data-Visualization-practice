```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(123)
comp_walk = []
for i in range(10000):
    steps = [0]
    for i in range(100):
        dice = np.random.randint(1,7)
        if dice <= 2:
            step = max(0, steps[-1] - 1)
        elif dice >= 3 and dice <= 5:
            step = steps[-1] + 1
        else:
            step = steps[-1] + np.random.randint(1,7)

        if np.random.rand() <= 0.001:
            step = 0;
        
        steps.append(step)
    comp_walk.append(steps)

comp_walk = np.transpose(np.array(comp_walk))
```

```python
type(comp_walk)
```

```python
plt.hist(comp_walk[-1, :], 50)
plt.title('Dice-Step Montecarlo Simulation')
plt.show()
```

```python
prob = np.sum(comp_walk[-1, :] >= 60)/10000.0
print(f'The probability that step 60 will be reached after 100 dice rolls is {prob * 100}')
```
