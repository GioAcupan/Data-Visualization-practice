```python
countries = pd.read_csv("countries (1).csv")
```

```python
# compare the population of philippines to japan
```

```python
ph = countries[countries.country == 'Philippines']
```

```python
jp = countries[countries.country == 'Japan']
```

```python
ph
```

```python
plt.plot(ph.year, ph.population / 10**6)
plt.plot(jp.year, jp.population / 10**6)

plt.xlabel("YEAR")
plt.ylabel("POPULATION (millions)")

plt.legend(["Philippines", "Japan"])
plt.show()
```

```python
#compute for the population growth per year of ph
ph_gpy_list = []
for i in range(ph.population.size):
    if i == 0:
        ph_gpy_list.append(None)
    else:
        ph_gpy_list.append((ph.population.iloc[i]/ph.population.iloc[i - 1]) * 100)
ph_gpy_series = pd.Series(ph_gpy_list)
#compute for the population growth per year of jp
jp_gpy_list = []
for i in range(jp.population.size):
    if i == 0:
        jp_gpy_list.append(None)
    else:
        jp_gpy_list.append((jp.population.iloc[i]/jp.population.iloc[i - 1]) * 100)
jp_gpy_series = pd.Series(jp_gpy_list)
    

plt.plot(ph.year, ph_gpy_series)
plt.plot(jp.year, jp_gpy_series)

plt.xlabel("YEAR")
plt.ylabel("% GROWTH PER YEAR")

plt.legend(["Philippines", "Japan"])
plt.show()
```
